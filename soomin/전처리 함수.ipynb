```python
potato = potato[['거래일자','품목명','반입량','금액','전년 반입량','전년 금액']]
potato
```

```python
# 날짜/품목 컬럼 추론
date_col_candidates = [c for c in potato.columns if "거래일자" in c or "date" in c.lower()]
if len(date_col_candidates) == 0:
    raise ValueError("`거래일자`(또는 date 유사) 컬럼을 찾을 수 없습니다.")
date_col = date_col_candidates[0]

item_col_candidates = [c for c in potato.columns if ("품목" in c) or ("품종" in c) or ("품목명" in c)]
item_col = item_col_candidates[0] if len(item_col_candidates) else "품목명"
if item_col not in df.columns:
    potato[item_col] = "미상"

# 날짜 파싱
potato[date_col] = pd.to_datetime(potato[date_col], errors="coerce")

def to_numeric_clean(s: pd.Series) -> pd.Series:
    return pd.to_numeric(
        s.astype(str)
         .str.replace(",", "", regex=False)
         .str.replace(" ", "", regex=False)
         .str.replace("%", "", regex=False)
         .str.replace("\u2212", "-", regex=False),  # 유니코드 음수 기호
        errors="coerce"
    )

num_candidates = [
    "반입량","금액",
    "전년 반입량",
    "전년 금액",
]
for c in num_candidates:
    if c in potato.columns:
        potato[c] = to_numeric_clean(potato[c])

# 파생: 연/월/요일/일요일/주차
potato["연도"] = potato[date_col].dt.year
potato["월"] = potato[date_col].dt.month
potato["weekday"] = potato[date_col].dt.weekday   # 0=월, 6=일
potato["is_sunday"] = (potato["weekday"] == 6)
potato["주차"] = potato[date_col].dt.isocalendar().week.astype("Int64")
potato
```

```python
import pandas as pd
from pandas.api.types import is_datetime64_any_dtype as is_datetime

potato_copy = potato.copy()

# 0) 날짜형 보장
if not is_datetime(potato_copy['거래일자']):
    potato_copy['거래일자'] = pd.to_datetime(potato_copy['거래일자'], errors='coerce')

# (옵션) 음수 금액은 결측으로 처리
potato_copy.loc[potato_copy['금액'] < 0, '금액'] = pd.NA

# 1) 전년도 후보 테이블 생성 (+1년 이동해서 "현재행 날짜"와 맞춤)
prev = potato_copy[['거래일자', '품목명', '반입량', '금액']].copy()
prev['거래일자'] = prev['거래일자'] + pd.DateOffset(years=1)
prev = prev.rename(columns={'반입량': '전년도_반입량_from_row', '금액': '전년도_금액_from_row'})

# 1-추가) 동일 키(거래일자+품목명) 중복 시 처리 (여기선 '마지막 값' 선택; 필요시 'first'로 변경)
prev = (
    prev.sort_values(['거래일자', '품목명'])
        .groupby(['거래일자', '품목명'], as_index=False)
        .last()
)

# 2) 병합
df = potato_copy.merge(prev, on=['거래일자', '품목명'], how='left', validate='m:1')

# 3) 채우기 (반입량/금액 각각 NaN인 곳만 전년도 행 값으로 보강)
before_nan_cnt_qty  = df['반입량'].isna().sum()
before_nan_cnt_amt  = df['금액'].isna().sum()

mask_any_nan = df['반입량'].isna() | df['금액'].isna()
filled_qty_mask = df['반입량'].isna() & df['전년도_반입량_from_row'].notna()
filled_amt_mask = df['금액'].isna() & df['전년도_금액_from_row'].notna()

df.loc[filled_qty_mask, '반입량'] = df.loc[filled_qty_mask, '전년도_반입량_from_row']
df.loc[filled_amt_mask, '금액']   = df.loc[filled_amt_mask, '전년도_금액_from_row']

after_nan_cnt_qty = df['반입량'].isna().sum()
after_nan_cnt_amt = df['금액'].isna().sum()

# 4) 채워진 건수/샘플 확인용
filled_rows_idx = df.index[filled_qty_mask | filled_amt_mask]
filled_count = len(filled_rows_idx)

print(f"[요약] 반입량 NaN: {before_nan_cnt_qty} → {after_nan_cnt_qty} (채움 {before_nan_cnt_qty - after_nan_cnt_qty})")
print(f"[요약] 금액   NaN: {before_nan_cnt_amt} → {after_nan_cnt_amt} (채움 {before_nan_cnt_amt - after_nan_cnt_amt})")
print(f"[요약] 전년도 행으로 보강된 총 행 수: {filled_count}")

# 5) 보조 컬럼 정리 (원하면 유지 가능)
df = df.drop(columns=['전년도_반입량_from_row', '전년도_금액_from_row'])
```
