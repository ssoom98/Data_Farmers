```python
# 1. 기타만 필터링
potato = potato_price[potato_price['품목명'] == '기타'].copy()

# 2. 날짜 타입 변환
potato['거래일자'] = pd.to_datetime(potato['거래일자'])

# 3. 월별 평균가격 / 반입량 집계
monthly_price = potato.resample('M', on='거래일자')['금액'].mean().reset_index()
monthly_amount = potato.resample('M', on='거래일자')['반입량'].mean().reset_index()

# 4. 이중축 라인 플롯
fig, ax1 = plt.subplots(figsize=(14,6))

# 좌측 y축 (평균금액)
ax1.plot(monthly_price['거래일자'], monthly_price['금액'], color='tab:blue', marker='o', label='평균금액')
ax1.set_xlabel('날짜')
ax1.set_ylabel('평균금액 (원)', color='tab:blue')
ax1.tick_params(axis='y', labelcolor='tab:blue')

# 우측 y축 (평균반입량)
ax2 = ax1.twinx()
ax2.plot(monthly_amount['거래일자'], monthly_amount['반입량'], color='tab:red', marker='s', label='평균반입량')
ax2.set_ylabel('평균반입량 (톤)', color='tab:red')
ax2.tick_params(axis='y', labelcolor='tab:red')

# 제목 & 격자
plt.title('기타 감자 월별 평균금액 & 평균반입량 추이')
fig.tight_layout()
plt.show()
```
