```python
# -*- coding: utf-8 -*-
# [ÏùÄÏ£ºÎãò Í≤∞Ï∏°Ïπò Ï≤òÎ¶¨ ÏãúÍ∞ÅÌôî] - ÌòÑÏû¨ Ìè¥Îçî(04 ÏùÄÏ£º)ÏóêÏÑú *.csv ÏùºÍ¥Ñ Î°úÎìú ‚Üí Í≤∞Ï∏° ÏãúÍ∞ÅÌôî
import os, re, glob
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from math import ceil

# -----------------------------
# 0) ÌôòÍ≤Ω (ÌïúÍ∏Ä Ìè∞Ìä∏/ÎßàÏù¥ÎÑàÏä§)
# -----------------------------
plt.rcParams["font.family"] = "AppleGothic"   # macOS
plt.rcParams["axes.unicode_minus"] = False

# -----------------------------
# 1) Í≤ΩÎ°ú/ÎåÄÏÉÅ ÌíàÎ™©
# -----------------------------
DATA_DIR = "./"  # ÌòÑÏû¨ ipynbÏôÄ Í∞ôÏùÄ Ìè¥Îçî(04 ÏùÄÏ£º)

TARGET_ITEMS = [
    "ÍπªÏûé","ÎØ∏ÎÇòÎ¶¨","Î∂ÄÏ∂î","Î∏åÎ°úÏΩúÎ¶¨",   # ‚úÖ Î∏åÎ°úÏΩúÎ¶¨Îßå ÏÇ¨Ïö© (Î∏åÎ°úÏΩîÎ¶¨Îäî Ïä§ÌÇµ)
    "ÏÉùÍ∞ï","ÏãúÍ∏àÏπò","Ïë•Í∞ì","Ïï†Ìò∏Î∞ï","ÏñëÎ∞∞Ï∂î",
    "ÏñëÏÉÅÏ∂î","Ïò§Ïù¥","Ï™ΩÌåå","ÏπòÏª§Î¶¨"
]

NUM_COLS = ['Í∏àÏï°','Î∞òÏûÖÎüâ','Ï†ÑÎÖÑ Í∏àÏï°','Ï†ÑÎÖÑ Î∞òÏûÖÎüâ','ÌèâÎÖÑ Í∏àÏï°','ÌèâÎÖÑ Î∞òÏûÖÎüâ']
DATE_CANDIDATES = ["Í±∞ÎûòÏùºÏûê","Í±∞ÎûòÏùº","ÏùºÏûê","ÎÇ†Ïßú","date","Date","DATE"]

# -----------------------------
# 2) Î°úÎî© Ïú†Ìã∏
# -----------------------------
def smart_read_csv(path):
    encodings = ["utf-8-sig", "cp949", "euc-kr", "utf-8"]
    last_err = None
    for enc in encodings:
        try:
            return pd.read_csv(path, encoding=enc)
        except Exception as e:
            last_err = e
    raise last_err

def to_numeric_safe(s):
    return pd.to_numeric(
        s.astype(str).str.replace(",", "", regex=False).str.strip(),
        errors="coerce"
    )

def pick_date_column(df):
    for c in DATE_CANDIDATES:
        if c in df.columns:
            return c
    return None

def extract_item_from_filename(fname: str) -> str:
    """'ÍπªÏûé.csv' ‚Üí ÍπªÏûé, 'Ï™ΩÌåågarakFix2025...' ‚Üí Ï™ΩÌåå"""
    stem = os.path.splitext(fname)[0]
    return stem.split("garakFix", 1)[0] if "garakFix" in stem else stem

# -----------------------------
# 3) CSV Î°úÎçî (Î∏åÎ°úÏΩîÎ¶¨ Ïä§ÌÇµ, Î∏åÎ°úÏΩúÎ¶¨Îßå)
# -----------------------------
def load_all_csvs_with_logs(data_dir, target_items):
    files = sorted(glob.glob(os.path.join(data_dir, "*.csv")))
    print(f"üîé CSV ÌÉêÏÉâ: {len(files)}Í∞ú")
    name_map = {}
    for f in files:
        fname = os.path.basename(f)
        raw_item = extract_item_from_filename(fname)

        # Ïò§ÏßÅ Î∏åÎ°úÏΩúÎ¶¨Îßå ÌóàÏö©
        if raw_item == "Î∏åÎ°úÏΩîÎ¶¨":
            print(f"‚ö† Ïä§ÌÇµ: {fname} (Î∏åÎ°úÏΩîÎ¶¨Îäî ÏÇ¨Ïö©ÌïòÏßÄ ÏïäÏùå)")
            continue

        item = raw_item
        if item not in target_items:
            # Ìè¥Îçî ÎÇ¥ Îã§Î•∏ CSVÎäî Î¨¥Ïãú
            continue

        # ÏùΩÍ∏∞
        try:
            df = smart_read_csv(f)
        except Exception as e:
            print(f"‚ùå ÏùΩÍ∏∞ Ïã§Ìå®: {fname} -> {e}")
            continue

        # ÎÇ†Ïßú Ïª¨Îüº
        date_col = pick_date_column(df)
        if date_col is None:
            print(f"‚ö† ÎÇ†Ïßú Ïª¨Îüº ÏóÜÏùå: {fname} / Ïª¨Îüº: {df.columns.tolist()}")
            continue

        df = df.copy()
        df[date_col] = pd.to_datetime(df[date_col], errors="coerce")

        # ÏàòÏπòÌòï Î≥ÄÌôò
        for col in NUM_COLS:
            if col in df.columns:
                df[col] = to_numeric_safe(df[col])

        # ÌëúÏ§ÄÌôî
        df.rename(columns={date_col: "Í±∞ÎûòÏùºÏûê"}, inplace=True)
        df["ÌíàÎ™©"] = item
        if "ÌíàÎ™©Î™Ö" not in df.columns:
            df["ÌíàÎ™©Î™Ö"] = item

        n_rows = len(df); n_valid = df["Í±∞ÎûòÏùºÏûê"].notna().sum()
        print(f"‚úÖ Î°úÎìú: {fname} | ÌíàÎ™©={item} | Ìñâ={n_rows} | Ïú†Ìö® ÎÇ†Ïßú={n_valid}")

        # Í∞ôÏùÄ ÌíàÎ™© Îã§Ïàò ÌååÏùº Ìï©ÏπòÍ∏∞
        name_map[item] = pd.concat([name_map[item], df], ignore_index=True) if item in name_map else df
    return name_map

# -----------------------------
# 4) Î°± Í≤∞Ìï© (ÏïàÏ†Ñ)
# -----------------------------
def build_long_df_safe(name_map):
    long_list = []
    for item, df in name_map.items():
        if "Í±∞ÎûòÏùºÏûê" not in df.columns:
            print(f"‚ö† Ïä§ÌÇµ({item}): 'Í±∞ÎûòÏùºÏûê' ÏóÜÏùå"); 
            continue
        tmp = df.dropna(subset=["Í±∞ÎûòÏùºÏûê"]).copy()
        if tmp.empty:
            print(f"‚ö† Ïä§ÌÇµ({item}): Ïú†Ìö® ÎÇ†Ïßú Ìñâ ÏóÜÏùå"); 
            continue
        tmp["Ïó∞ÎèÑ"] = tmp["Í±∞ÎûòÏùºÏûê"].dt.year
        long_list.append(tmp)
    if not long_list:
        raise ValueError("Í≤∞Ìï©Ìï† Îç∞Ïù¥ÌÑ∞Í∞Ä ÏóÜÏäµÎãàÎã§. Î°úÎî© Î°úÍ∑∏Î•º ÌôïÏù∏ÌïòÏÑ∏Ïöî.")
    long_df = pd.concat(long_list, ignore_index=True)
    use_cols = ['Í±∞ÎûòÏùºÏûê','ÌíàÎ™©','ÌíàÎ™©Î™Ö','Ïó∞ÎèÑ'] + [c for c in ['Í∏àÏï°','Î∞òÏûÖÎüâ'] if c in long_df.columns]
    return long_df[use_cols].copy()

# -----------------------------
# 5) ÏãúÍ∞ÅÌôî Ïú†Ìã∏
# -----------------------------
def _prep_item(df_all, target_item):
    return df_all[df_all['ÌíàÎ™©'] == target_item].copy()

def plot_subitem_year_missing_heatmap_for_item(df_all, target_item, value_col, last_n_years=10):
    df = _prep_item(df_all, target_item)
    if value_col not in df.columns:
        print(f"‚ö† {target_item}: '{value_col}' ÏóÜÏùå"); return
    grp = (df.assign(Í≤∞Ï∏°=df[value_col].isna().astype(int))
             .groupby(['ÌíàÎ™©Î™Ö','Ïó∞ÎèÑ'], as_index=False)['Í≤∞Ï∏°'].mean())
    if grp.empty:
        print(f"‚ö† {target_item}: ÏßëÍ≥Ñ Îç∞Ïù¥ÌÑ∞ ÏóÜÏùå"); return
    grp['Í≤∞Ï∏°Î•†(%)'] = grp['Í≤∞Ï∏°'] * 100
    years = sorted(grp['Ïó∞ÎèÑ'].dropna().unique())
    if last_n_years and len(years) > last_n_years:
        years = years[-last_n_years:]
    grp = grp[grp['Ïó∞ÎèÑ'].isin(years)]
    if grp.empty:
        print(f"‚ö† {target_item}: ÏµúÍ∑º {last_n_years}ÎÖÑ Îç∞Ïù¥ÌÑ∞ ÏóÜÏùå"); return
    pivot_tbl = grp.pivot(index='ÌíàÎ™©Î™Ö', columns='Ïó∞ÎèÑ', values='Í≤∞Ï∏°Î•†(%)')
    fig, ax = plt.subplots(figsize=(1 + 0.4*len(years), 1 + 0.45*len(pivot_tbl.index)))
    c = ax.pcolor(pivot_tbl.values, cmap='Greys')  # Í≤Ä=ÎÇÆÏùå, Ìù∞=ÎÜíÏùå
    fig.colorbar(c, ax=ax, label='Í≤∞Ï∏°Î•†(%)')
    ax.set_xticks(np.arange(pivot_tbl.shape[1]) + 0.5)
    ax.set_xticklabels(pivot_tbl.columns, rotation=45, ha='right')
    ax.set_yticks(np.arange(pivot_tbl.shape[0]) + 0.5)
    ax.set_yticklabels(pivot_tbl.index)
    ax.set_title(f"[{target_item}] {value_col} Í≤∞Ï∏°Î•† ÌûàÌä∏Îßµ")
    plt.tight_layout(); plt.show()

def plot_subitem_year_bars_for_item(df_all, target_item, value_col, last_n_years=10):
    df = _prep_item(df_all, target_item)
    if value_col not in df.columns:
        print(f"‚ö† {target_item}: '{value_col}' ÏóÜÏùå"); return
    sub_items = sorted(df['ÌíàÎ™©Î™Ö'].dropna().unique())
    if not sub_items:
        print(f"‚ö† {target_item}: ÏÑ∏Î∂Ä ÌíàÎ™©Î™Ö ÏóÜÏùå"); return

    n = len(sub_items)
    ncols = 3 if n >= 7 else (2 if n >= 3 else 1)
    nrows = int(np.ceil(n / ncols))
    fig, axes = plt.subplots(nrows=nrows, ncols=ncols, figsize=(4.6*ncols, 3.8*nrows))

    # ‚úÖ Îã®Ïùº Axes ÎåÄÎπÑ
    axes = axes.flatten() if isinstance(axes, np.ndarray) else [axes]

    for i, sub_name in enumerate(sub_items):
        ax = axes[i]
        ser = (df[df['ÌíàÎ™©Î™Ö']==sub_name]
               .groupby('Ïó∞ÎèÑ')[value_col]
               .apply(lambda x: x.isna().mean()*100)
               .sort_index())
        if last_n_years and len(ser.index) > last_n_years:
            ser = ser.iloc[-last_n_years:]
        ax.bar(ser.index.astype(str), ser.values)
        ax.set_title(f"{sub_name} - {value_col}")
        ax.set_ylim(0, 100)
        ax.set_xlabel("Ïó∞ÎèÑ"); ax.set_ylabel("Í≤∞Ï∏°Î•†(%)")
        ax.set_xticklabels(ser.index.astype(str), rotation=45, ha='right')

    # ÎÇ®Îäî Ï∂ï Ï†úÍ±∞
    for j in range(i+1, len(axes)):
        try:
            fig.delaxes(axes[j])
        except Exception:
            pass

    fig.suptitle(f"[{target_item}] {value_col} Í≤∞Ï∏°Î•†(%)", y=1.02)
    plt.tight_layout(); plt.show()

def run_missing_visuals_for_item(df_all, target_item, last_n_years=10):
    print(f"\n=== {target_item} Í≤∞Ï∏° ÏãúÍ∞ÅÌôî ===")
    for col in ["Í∏àÏï°","Î∞òÏûÖÎüâ"]:
        plot_subitem_year_missing_heatmap_for_item(df_all, target_item, col, last_n_years)
    for col in ["Í∏àÏï°","Î∞òÏûÖÎüâ"]:
        plot_subitem_year_bars_for_item(df_all, target_item, col, last_n_years)

# -----------------------------
# 6) Ïã§Ìñâ
# -----------------------------
name_map = load_all_csvs_with_logs(DATA_DIR, TARGET_ITEMS)
if not name_map:
    raise RuntimeError("ÎåÄÏÉÅ CSVÍ∞Ä ÌïòÎÇòÎèÑ Î°úÎìúÎêòÏßÄ ÏïäÏïòÏäµÎãàÎã§. Î°úÎî© Î°úÍ∑∏Î•º ÌôïÏù∏ÌïòÏÑ∏Ïöî.")

df_use = build_long_df_safe(name_map)

for it in name_map.keys():      # Ïã§Ï†úÎ°ú ÏùΩÌûå ÌíàÎ™©Îßå Ïã§Ìñâ
    run_missing_visuals_for_item(df_use, it, last_n_years=10)
```
